```python
#Câu 1: Viết hàm để Tính diện tích tam giác
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam giác không hợp lệ")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Diện tích =",dt)
```

```python
#Câu 3: Viết Hàm tính BMI
def BMI(height,weight):
 return weight/(height**2)
def PhanLoai(bmi):
 if bmi<18.5:
    return "Gầy"
 elif bmi<=24.9:
    return "Bình thường"
 elif bmi<=29.9:
    return "Hơi Béo"
 elif bmi<=34.9:
    return "Béo Phì Cấp Độ 1"
 elif bmi<=39.9:
    return "Béo Phì Cấp Độ 2"
 else:
    return "Béo Phì Cấp độ 3"
def NguyCoBenh(bmi):
    if bmi<18.5:
        return "Thấp"
    elif bmi<=24.9:
        return "Trung Bình"
    elif bmi<=29.9:
        return "Cao"
    elif bmi<=34.9:
        return "Cao"
    elif bmi<=39.9:
        return "Rất cao"
    else:
        return "Nguy Hiểm"
print("Nhập vào chiều cao:")
height=float(input())
print("Nhập vào cân nặng:")
weight=float(input())
bmi=BMI(height,weight)
print("BMI của bạn=",bmi)
print("Phân loại bạn=",PhanLoai(bmi))
print("Nguy cơ bệnh của Thím=",NguyCoBenh(bmi))
```

```python
#Câu 4: Viết Hàm tính ROI
def ROI(dt,cp):
 return (dt-cp)/cp
def GoiYDauTu(roi):
 if roi>=0.75:
    return "Nên đầu tư"
 else:
    return "Không nên đầu tư"
print("Chương trình tính ROI")
dt=int(input("Nhập Doanh Thu:"))
cp=int(input("Nhập chi phí:"))
roi=ROI(dt,cp)
print("Tỉ Lệ ROI=",roi)
print("==>",GoiYDauTu(roi))
```

```python
#Câu 5: Viết hàm đệ qui Fibonacci
def fibonacci(n):
 if n<=2 :
    return 1
 return fibonacci(n-1)+fibonacci(n-2)
def listfibo(n):
 for i in range(1,n+1):
    print(fibonacci(i),end='\t')
print(fibonacci(9))
listfibo(9)
```

```python
#Câu 7: Tính và xuất độ dài đoạn AB
import math

# Nhập tọa độ 2 điểm
xA = float(input("Nhập xA: "))
yA = float(input("Nhập yA: "))
xB = float(input("Nhập xB: "))
yB = float(input("Nhập yB: "))

# Công thức tính độ dài AB
dAB = math.sqrt((xB - xA)**2 + (yB - yA)**2)

print(f"Độ dài đoạn AB = {dAB}")
```

```python
#Câu 12: Hàm oscillate
def oscillate(a, b):
    n = a
    step = 1
    while n <= b:
        yield n
        n += step
        if n == 1:      # khi tới 1 thì quay đầu đi xuống
            step = -1
        elif n == a:    # khi về tới a thì đổi chiều đi lên
            step = 1

# Test
for n in oscillate(-3, 5):
    print(n, end=" ")
```

```python
#Câu 13: Hàm kiểm tra số hoàn thiện, số thịnh vượng
# Hàm tính tổng các ước số (không tính chính nó)
def tong_uoc(n):
    tong = 1   # luôn có ước số 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            tong += i
            if i != n // i:  # tránh cộng trùng khi i*i = n
                tong += n // i
    return tong if n != 1 else 0   # số 1 không có ước số nào

# Hàm kiểm tra số hoàn thiện
def la_so_hoan_thien(n):
    return tong_uoc(n) == n

# Hàm kiểm tra số thịnh vượng
def la_so_thinh_vuong(n):
    return tong_uoc(n) > n

# --- Test ---
n = int(input("Nhập số nguyên dương n: "))

if la_so_hoan_thien(n):
    print(f"{n} là số hoàn thiện")
elif la_so_thinh_vuong(n):
    print(f"{n} là số thịnh vượng")
else:
    print(f"{n} không phải số hoàn thiện cũng không phải số thịnh vượng")
```
