```python
#Câu 1: Xử lý List
from random import randrange

print("Chương trình xử lý List")
n = int(input("Nhập số phần tử: "))
lst = [0]*n

# Sinh list ngẫu nhiên
for i in range(n):
    lst[i] = randrange(-100, 100)

print("List ngẫu nhiên là:")
print(lst)

# Thêm số mới vào list
print("Mời bạn thêm số mới:")
value = int(input())
lst.append(value)
print(lst)

# Đếm số xuất hiện
print("Bạn muốn đếm số nào:")
k = int(input())
dem = lst.count(k)
print(k, "xuất hiện", dem, "lần trong list")

# Hàm kiểm tra số nguyên tố
def CheckPrime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Đếm & tính tổng số nguyên tố
demnt = 0
tongnt = 0
for x in lst:
    if CheckPrime(x):
        demnt += 1
        tongnt += x

print("Có", demnt, "số nguyên tố trong list")
print("Tổng =", tongnt)

# Sắp xếp list
lst.sort()
print("List sau khi sort:")
print(lst)

# Xóa list → làm rỗng thay vì del
lst.clear()
print("List sau khi xóa:")
print(lst)
```

```python
#Câu 3: Xử lý List Đa chiều
from random import randrange
def TaoMaTran(m,n):
 D=[]
 for i in range(m):
    row=[]
    for j in range(n):
        row.append(randrange(100))
    D.append(row)
 return D
def XuatMaTran(D):
 for row in D:
    for element in row:
        print(element,end='\t')
    print()
def LayDong(r):
 R=D[r]
 return R
def XuatList1Chieu(R):
 for element in R:
    print(element,end='\t')
def LayCot(c):
 C=[]
 for i in range(len(D)):
    C.append(D[i][c])
 return C
def MAX(D):
 max=D[0][0]
 for i in range(len(D)):
    for j in range(len(D[i])):
        if(max<D[i][j]):
            max=D[i][j]
 return max
print("Nhập số dòng:")
m=int(input())
print("Nhập số cột:")
n=int(input())
D=TaoMaTran(m,n)
XuatMaTran(D)
print("Mời bạn nhập dòng muốn xuất:")
r=int(input())
XuatList1Chieu(LayDong(r))
print("Mời bạn nhập cột muốn xuất:")
c=int(input())
XuatList1Chieu(LayCot(c))
max=MAX(D)
print("Số lớn nhất trong ma trận=",max)
```

```python
#Câu 4: Xác định kết quả khi thực thi list
lst = [3, 0, 1, 5, 2]
x = 2
print(lst[0])        # (a) → 3
print(lst[x])        # (b) → lst[2] = 1
print(lst[-1])       # (c) → 2 (phần tử cuối)
print(lst[x+1])      # (d) → lst[3] = 5
print(lst[-x])       # (e) → lst[-2] = 5
print(lst[-x+1])     # (f) → lst[-1] = 2
print(lst[1:3])      # (g) → [0, 1]
print(lst[x:-1])     # (h) → lst[2:-1] = [1, 5]
print(lst[::2])      # (i) → [3, 1, 2]
print(lst[x::2])     # (j) → [1, 2]
```

```python
#Câu 5: Xác định kết quả khi thực thi list

lst = [20, 1, -34, 40, -8, 60, 1, 3]
print(lst)         # (a) → [20, 1, -34, 40, -8, 60, 1, 3]
print(lst[0:3])    # (b) → [20, 1, -34]
print(lst[4:8])    # (c) → [-8, 60, 1, 3]
print(lst[4:33])   # (d) → [-8, 60, 1, 3] (chỉ lấy đến hết list)
print(lst[-5:3])   # (e) → [] (vì -5=40, index 3 = 40, cắt xuôi không hợp lệ)
print(lst[3:-1])   # (f) → [40, -8, 60, 1]
print(lst[::-1])   # (g) → [3, 1, 60, -8, 40, -34, 1, 20] (đảo ngược list)
print(lst[:])      # (h) → [20, 1, -34, 40, -8, 60, 1, 3] (copy nguyên list)
print(lst[::])     # (i) → [20, 1, -34, 40, -8, 60, 1, 3] (copy nguyên list)
print(lst[::3])    # (j) → [20, 40, 1] (cứ bước 3 phần tử)
print(-34 in lst)  # (k) → True
print(34 not in lst) # (l) → True (vì list có -34, không có 34)
print(len(lst))    # (m) → 8
```

```python
#Câu 6: Nhập vào 1 list có N số ngẫu nhiên KHÔNG TRÙNG NHAU
def nhap_list_khong_trung():
    n = int(input("Nhập số phần tử N: "))
    lst = []
    while len(lst) < n:
        x = int(input(f"Nhập phần tử thứ {len(lst)+1}: "))
        if x not in lst:
            lst.append(x)
        else:
            print("⚠️ Bị trùng, nhập lại!")
    return lst

lst = nhap_list_khong_trung()
print("List sau khi nhập:", lst)
```

```python
#Câu 7: Viết chương trình nhập vào một dãy các số theo thứ tự tăng, nếu nhập sai quy cách thì yêu cầu nhập lại. In dãy số sau khi đã nhập xong 
def nhap_day_tang():
    n = int(input("Nhập số phần tử: "))
    lst = []
    for i in range(n):
        x = int(input(f"Nhập phần tử thứ {i+1}: "))
        if i > 0 and x <= lst[-1]:
            print("⚠️ Sai quy tắc tăng dần!")
            return []
        lst.append(x)
    return lst

lst = nhap_day_tang()
if lst:
    print("Dãy tăng dần:", lst)
```

```python
#Câu 8: Nhập dãy M[0], M[1], … M[n-1], sắp xếp giảm dần
def sap_xep_giam():
    n = int(input("Nhập số phần tử: "))
    M = []
    for i in range(n):
        M.append(int(input(f"M[{i}] = ")))
    print("Dãy ban đầu:", M)
    M.sort(reverse=True)
    print("Dãy sau khi sắp xếp giảm dần:", M)

sap_xep_giam()
```

```python
#Câu 9: Xử lý mảng số nguyên
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

M = [3,7,11,17,20,92,65,43]

print("Mảng:", M)
print("Số phần tử:", len(M))
print("Tổng:", sum(M))
print("Trung bình cộng:", sum(M)/len(M))
print("Các số chẵn:", [x for x in M if x % 2 == 0])
print("Các số lẻ:", [x for x in M if x % 2 != 0])
print("Các số nguyên tố:", [x for x in M if is_prime(x)])
print("Các số > số kế cuối:", [x for x in M if x > M[-2]])
```

```python
#Câu 10: Xử lý ma trận (cộng, trừ, nhân)
def nhap_matrix(m, n, name="A"):
    print(f"Nhập ma trận {name} ({m}x{n}):")
    mat = []
    for i in range(m):
        row = list(map(int, input(f"Hàng {i+1}: ").split()))
        while len(row) != n:
            print("⚠️ Sai số cột, nhập lại!")
            row = list(map(int, input(f"Hàng {i+1}: ").split()))
        mat.append(row)
    return mat

def cong_matrix(A, B):
    m, n = len(A), len(A[0])
    C = [[A[i][j] + B[i][j] for j in range(n)] for i in range(m)]
    return C

def nhan_matrix(A, B):
    m, n = len(A), len(A[0])
    n2, p = len(B), len(B[0])
    if n != n2:
        print("⚠️ Không thể nhân, số cột A phải bằng số hàng B")
        return None
    C = [[0]*p for _ in range(m)]
    for i in range(m):
        for j in range(p):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

def xuat_matrix(M, name="KQ"):
    print(f"Ma trận {name}:")
    for row in M:
        print(row)

# ====== MAIN ======
m, n = map(int, input("Nhập kích thước ma trận A (m n): ").split())
A = nhap_matrix(m, n, "A")

m2, n2 = map(int, input("Nhập kích thước ma trận B (m n): ").split())
B = nhap_matrix(m2, n2, "B")

xuat_matrix(A, "A")
xuat_matrix(B, "B")

if m == m2 and n == n2:
    C = cong_matrix(A, B)
    xuat_matrix(C, "A + B")
else:
    print("⚠️ Không cộng được vì kích thước A và B khác nhau")

D = nhan_matrix(A, B)
if D:
    xuat_matrix(D, "A x B")
```
